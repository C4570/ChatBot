```python
import csv
from transformers import BertModel, BertTokenizer
import torch
import numpy as np

# Cargar el modelo BERT y el tokenizador
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def get_bert_embedding(text):
    # Tokenizar el texto y obtener los embeddings
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)

    # Usar el embedding del token [CLS] como representación de la oración
    embeddings = outputs.last_hidden_state[:, 0, :].numpy()
    return embeddings.tolist()[0]

ejercicios = [
    {
        "Unidad": 1,
        "Tema": "Codificación de archivos",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Investigue los archivos en la carpeta 'Ejemplos codificación' con las herramientas vistas en clase. Use también hexdump para verlos por dentro. Pruebe abrir el archivo UTF-8 como 8859-15 y compare las salidas.",
        "Dificultad": "Media"
    },
    {
        "Unidad": 1,
        "Tema": "Extracción de texto",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Extraiga información relevante de archivos en diferentes formatos (pdf, jpg, etc.) utilizando las librerías desarrolladas en clase.",
        "Dificultad": "Alta"
    },
    {
        "Unidad": 1,
        "Tema": "OCR",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Obtenga texto de un libro escaneado en pdf usando pytesseract en inglés y en español.",
        "Dificultad": "Alta"
    },
    {
        "Unidad": 1,
        "Tema": "Extracción de texto de imagen",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Obtenga texto de una imagen (png, bmp).",
        "Dificultad": "Media"
    },
    {
        "Unidad": 1,
        "Tema": "Extracción de texto de Word",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Obtenga texto de un archivo word.",
        "Dificultad": "Fácil"
    },
    {
        "Unidad": 1,
        "Tema": "Extracción de texto de audio",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Obtenga texto de un archivo de audio.",
        "Dificultad": "Alta"
    },
    {
        "Unidad": 1,
        "Tema": "Web Scraping",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Extraiga los valores del tablero de indicadores de Objetivos de Desarrollo Sostenible de tableros.yvera.tur.ar en una tabla y el texto limpio de la metodología.",
        "Dificultad": "Media"
    },
    {
        "Unidad": 1,
        "Tema": "Web Scraping",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Extraiga información de vuelo en flybondi.com y verifique si bajó el precio de un vuelo determinado.",
        "Dificultad": "Media"
    },
    {
        "Unidad": 1,
        "Tema": "Web Scraping y Análisis",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Recolecte información diaria del tipo de cambio oficial de dolarhoy.com y realice un análisis visual y estadístico.",
        "Dificultad": "Alta"
    },
    {
        "Unidad": 1,
        "Tema": "Extracción de PDF y Serialización",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Obtenga los libros en formato pdf de la editorial Orsai de Hernán Casciari y guarde el texto en un archivo pkl.",
        "Dificultad": "Media"
    },
    {
        "Unidad": 1,
        "Tema": "Web Scraping y Procesamiento de Texto",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Obtenga el texto de la página de Wikipedia sobre Argentina, aplique diferentes métodos de splitting y compare los resultados.",
        "Dificultad": "Alta"
    },
    {
        "Unidad": 1,
        "Tema": "Extracción de Texto Avanzada",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Practique la extracción de texto de recursos desafiantes proporcionados en una carpeta de Google Drive.",
        "Dificultad": "Alta"
    },
    # Unidad 2
    {
        "Unidad": 2,
        "Tema": "One-Hot Encoding",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Escriba ejemplos de frases o busque ejemplos de párrafos de texto con diferentes estructuras, palabras y signos de puntuación. Utilice los códigos de One-Hot encoding en las dos versiones que presenta el material expuesto en clase. Analice los resultados.",
        "Dificultad": "Media"
    },
    {
        "Unidad": 2,
        "Tema": "Vectorización de Texto",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Use CountVectorizer y TfidfVectorizer para representar los documentos 'PROYECTO DE LEY TURISMO SOCIAL 2004.pdf' y el texto de la metodología del Ministerio de Turismo en forma de matrices numéricas. Compare las diferencias entre las dos representaciones.",
        "Dificultad": "Alta"
    },
    {
        "Unidad": 2,
        "Tema": "HashingVectorizer",
        "Tipo_Ejercicio": "Práctico y Conceptual",
        "Enunciado": "Utilice HashingVectorizer en los documentos anteriores. Explique qué significa que esta metodología es 'sin estado', sus ventajas, problemas potenciales, el concepto de 'matriz dispersa', y el rango numérico de los valores con norm=None.",
        "Dificultad": "Alta"
    },
    {
        "Unidad": 2,
        "Tema": "Semántica y Contexto",
        "Tipo_Ejercicio": "Conceptual",
        "Enunciado": "Explique brevemente por qué los métodos utilizados en los ejercicios anteriores no capturan la semántica y el contexto de las palabras.",
        "Dificultad": "Media"
    },
    {
        "Unidad": 2,
        "Tema": "Similaridad de Coseno",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Genere un código en Python que tome el resultado de una matriz de vectores de los ejercicios anteriores y extraiga las 10 similaridades de coseno mayores, imprimiendo las palabras correspondientes cuando sea posible.",
        "Dificultad": "Alta"
    },
    {
        "Unidad": 2,
        "Tema": "Word2Vec",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Cargue el modelo Word2Vec y explore palabras y similitudes en palabras de su interés. Busque posibilidades y combinaciones donde el uso de la aritmética de palabras dé resultados semánticamente coherentes.",
        "Dificultad": "Media"
    },
    {
        "Unidad": 2,
        "Tema": "Comparación de Modelos de Embeddings",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Utilice el texto recuperado del autor Hernán Casciari para crear un corpus de texto, elimine las stopwords y utilice Word2Vec, GloVe y FastText. Compare los resultados.",
        "Dificultad": "Alta"
    },
    {
        "Unidad": 2,
        "Tema": "Doc2Vec",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Separe por párrafos u oraciones el texto recopilado de Casciari. Use doc2vec sobre este corpus.",
        "Dificultad": "Alta"
    },
    # Unidad 3
     {
        "Unidad": 3,
        "Tema": "POS Tagging",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Utilice las librerías spacy y stanza para realizar un POS sobre los textos dados. Genere una función que extraiga la cantidad de sustantivos y verbos encontrados. Compare los resultados de ambas librerías.",
        "Dificultad": "Media"
    },
    {
        "Unidad": 3,
        "Tema": "Web Scraping y Clasificación de Texto",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Genere un web scraping de 5 categorías de frases de proverbia.net. Entrene un clasificador usando sklearn con los datos vectorizados. Pruebe diferentes modelos de embeddings, obtenga métricas y grafique las palabras más influyentes.",
        "Dificultad": "Alta"
    },
    {
        "Unidad": 3,
        "Tema": "Extracción de Entidades Nombradas y Resumen de Texto",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Extraiga texto de los archivos proporcionados, aplique un modelo para extraer Entidades Nombradas relevantes y genere un resumen de cada texto. Analice los resultados con diferentes modelos y parametrizaciones.",
        "Dificultad": "Alta"
    },
    {
        "Unidad": 3,
        "Tema": "Análisis de Sentimientos",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Explore los textos de reseñas de un producto usando técnicas de análisis de sentimientos. Analice el valor de cada comentario y genere estadísticas principales del conjunto. Utilice análisis gráfico para la resolución.",
        "Dificultad": "Media"
    },

    # Unidad 5
    {
        "Unidad": 5,
        "Tema": "Vectorización y Búsqueda de Similitud",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Use títulos de libros como corpus. Evalúe distancia para cada categoría usando KNN y el algoritmo Annoy. Compare resultados.",
        "Dificultad": "Alta"
    },
    {
        "Unidad": 5,
        "Tema": "Base de Datos Vectorial",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Almacene datos de noticias en ChromaDB. Explore categorías y palabras frecuentes. Genere índices y realice consultas. Compare resultados usando títulos y contenidos.",
        "Dificultad": "Alta"
    },
    {
        "Unidad": 5,
        "Tema": "Chunking y Almacenamiento en ChromaDB",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Divida el texto de Hernán Casciari en chunks usando RecursiveCharacterTextSplitter. Almacene en ChromaDB, realice búsquedas y analice problemas encontrados.",
        "Dificultad": "Media"
    },
    {
        "Unidad": 5,
        "Tema": "Consultas a Base de Datos de Grafos",
        "Tipo_Ejercicio": "Práctico",
        "Enunciado": "Investigue ejemplos de Query service en Wikidata. Use SPARQLWrapper para consultas a base de datos de grafos. Opcional: genere triadas para contextualizar a chatGPT.",
        "Dificultad": "Alta"
    }
]

filename = "ejercicios_nlp_con_embeddings.csv"

with open(filename, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=["Unidad", "Tema", "Tipo_Ejercicio", "Enunciado", "Dificultad", "Embedding"])
    writer.writeheader()
    for ejercicio in ejercicios:
        # Calcular el embedding del enunciado
        embedding = get_bert_embedding(ejercicio["Enunciado"])
        ejercicio["Embedding"] = embedding
        writer.writerow(ejercicio)

print(f"CSV creado con embeddings: {filename}")
```

### Visualizar el grafo

```python
pos = nx.spring_layout(G, k=0.5, iterations=50)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=4000, font_size=7, font_weight='bold')
edge_labels = nx.get_edge_attributes(G, 'relacion')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=6)

# Añadir las definiciones como etiquetas
node_labels = nx.get_node_attributes(G, 'definicion')
pos_attrs = {}
for node, coords in pos.items():
    pos_attrs[node] = (coords[0], coords[1] - 0.1)

nx.draw_networkx_labels(G, pos_attrs, labels=node_labels, font_size=5, font_weight='normal')

plt.title("Grafo de conceptos de Extracción y Procesamiento de Texto")
plt.axis('off')
plt.tight_layout()
plt.figure(figsize=(20, 8))
plt.show()
```
